```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
import sklearn 
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
```

```python
#Here is the data file I'm using. It's a data set about customers using an ecommerce platform.
customers = pd.read_csv("Ecommerce Customers")
```

```python
#Here is relevant metadata and information about each customer. 
#There are 500 customers recorded, which is a big enough sample size to get a good understanding of 
# what customer's want, provided the data is accurate.
customers.info()
```

```python
#Below are two Joint Plots in a Hex format: 
```

```python

sns.jointplot(customers['Time on App'],customers['Yearly Amount Spent'],kind='hex',color='red')
sns.jointplot(customers['Time on Website'],customers['Yearly Amount Spent'],kind='hex',color='orange')

print("The darker shades show where most customers fall under, in terms of their time spent on the platform")
print("And how much they have spent in year on the service")

print("This is a great way of seeing the choices customers are making while using your platform.")

#
```

```python
# Here is a line of regression: it shows a strong positive correlation
# between how long a customer has been a member for and how much they spend per year.
print("This line of regression shows a strong positive correlation")
print("between how long a customer has been a member for and how much they spend per year.")

graph = sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=customers,line_kws={'color': 'red'})
```

```python
# Here I make two varaible sets from the Customer Dataset.
# For 

X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']

#Here, I have split the Data into a training set, 
# which will be used to build up my Linear Regression model.
# and into a test set, which will show how accurately the model can make predictions.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = LinearRegression()
# I created a model and trained it below using the values provided from the customer dataset.
lm.fit(X_train,y_train)
```

```python
lm.predict(X_test)
plt.scatter(lm.predict(X_test),y_test,color='red')
plt.xlabel("The predicted values for Y:")
plt.ylabel("The actual values for Y:")
```
